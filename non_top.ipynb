```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0'
}

def scrape_position(year, position, ranks):
    url = f'https://247sports.com/season/{year}-football/TransferPortalTop/?positionKey={position}'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    players = soup.find_all('li', class_='transfer-player')
    data = {'First': [], 'Last': [], 'Height': [], 'Weight': [], 'Score': []}
    for player in players:
        rank_tag = player.find('div', class_='playerRank')
        if rank_tag:
            rank = int(rank_tag.text.strip())
        else:
            continue
        if rank not in ranks:
            continue
        name_tag = player.find('h3').find('a')
        if name_tag:
            name = name_tag.text.strip().split()
            first = name[0]
            last = name[1]
        else:
            first = last = None
        
        bio_tag = player.find('div', class_='bio')
        if bio_tag:
            hw = bio_tag.text.strip().replace('\n','').split('/')
            if len(hw) == 2:
                height = hw[0].strip()
                weight = hw[1].strip()
            else:
                height = weight = None
        else:
            height = weight = None
        score_tag = player.find('div', class_='rating')
        score = score_tag.text.strip() if score_tag else None
        data['First'].append(first)
        data['Last'].append(last)
        data['Height'].append(height)
        data['Weight'].append(weight)
        data['Score'].append(score)
    return data
```

```python
config = {
    2023: {
        '24': [101,104],
        '61': [56,94,42,66],
        '60': [29],
        '58': [23,56,51],
        '59': [36],
        '16': [50,77,22,28,32],
        '57': [19,26,21],
        '12': [54,32],
        '15': [18],
        '14': [125]
    },
    2024: {
        '24': [104],
        '58': [108,67],
        '57': [76,84,57],
    },
    2025: {
        '61': [76,144,209,240],
        '60': [98],
        '58': [54],
        '59': [148],
        '16': [45],
        '57': [94],
        '12': [114],
        '68': [22,154,99],
        '14': [63]
    },
    2026: {
        '24': [3,176],
        '59': [126],
        '58': [63,23],
        '16': [134],
        '68': [17],
        '14': [58,144]
    }
}
```

```python
all_data = {'First': [], 'Last': [], 'Year': [], 'Height': [], 'Weight': [], 'Score': []}

for year in config:
    for position in config[year]:
        
        ranks = config[year][position]
        
        temp = scrape_position(year, position, ranks)
        
        for i in range(len(temp['First'])):
            all_data['First'].append(temp['First'][i])
            all_data['Last'].append(temp['Last'][i])
            all_data['Year'].append(year)
            all_data['Height'].append(temp['Height'][i])
            all_data['Weight'].append(temp['Weight'][i])
            all_data['Score'].append(temp['Score'][i])
```

```python
non_top_transfer = pd.DataFrame(all_data)
non_top_transfer['Height'] = non_top_transfer['Height'].apply(
    lambda x: int(x.split('-')[0]) * 12 + int(x.split('-')[1]) if pd.notnull(x) else None
)
non_top_transfer = non_top_transfer.sort_values(by = ['Year','Last'])
print(non_top_transfer.head())
non_top_transfer.to_csv('transfer_datasets/non_top_transfer.csv', index=False)
```
